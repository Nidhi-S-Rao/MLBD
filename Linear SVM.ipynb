```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
from google.colab import drive
drive.mount('/content/drive/')
```

```python
data=pd.read_csv('drive/My Drive/Colab Notebooks/Social_Network_Ads.csv')
```

```python
data
```

```python
x=data.iloc[:,1:4].values
```

```python
x
```

```python
y=data.iloc[:,4].values
```

```python
y
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder_X = LabelEncoder()
x[:, 0] = labelencoder_X.fit_transform(x[:, 0])
x
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Standard scalar
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train=st.fit_transform(X_train)
x_test=st.transform(X_test)
```

```python
linear_svm_classifier = SVC(kernel='linear',random_state=0)
linear_svm_classifier.fit(x_train, y_train)
```

```python
y_pred_linear = linear_svm_classifier.predict(x_test)
```

```python
accuracy_linear = accuracy_score(y_test, y_pred_linear)
print(f"Linear SVM Accuracy: ",accuracy_linear*100,"%")
```

```python
data=pd.read_csv('drive/My Drive/Colab Notebooks/Social_Network_Ads.csv')
```

```python
x=data.iloc[:,2:4].values
```

```python
y=data.iloc[:,4].values
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
#Standard scalar
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train=st.fit_transform(X_train)
x_test=st.transform(X_test)
```

```python
linear_svm_classifier = SVC(kernel='linear',random_state=0)
linear_svm_classifier.fit(x_train, y_train)
```

```python
y_pred_linear = linear_svm_classifier.predict(x_test)
```

```python
accuracy_linear = accuracy_score(y_test, y_pred_linear)
print(f"Linear SVM Accuracy: ",accuracy_linear*100,"%")
```

```python

import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap

X_set , Y_set = x_train ,y_train
X1, X2 = np.meshgrid(np.arange(start = X_set[:,0].min()-1, stop = X_set[:,0].max()+1 , step = 0.01),
                     np.arange(start = X_set[:,1].min()-1, stop = X_set[:,1].max()+1 , step = 0.01))
plt.contourf(X1 , X2, linear_svm_classifier.predict(np.array([X1.ravel(),X2.ravel()]).T).reshape(X1.shape),alpha = 0.75,cmap =ListedColormap(('green','red')))
plt.xlim(X1.min(),X1.max())
plt.xlim(X2.min(),X2.max())

for i, j in enumerate(np.unique(Y_set)):
  plt.scatter(X_set[Y_set == j,0],X_set[Y_set ==j,1],
              c = ListedColormap(('green','red'))(i),label =j)
plt.title('SVM(Training Set)')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.legend()
plt.show()
```
