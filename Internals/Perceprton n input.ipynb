```python
def and_gate(inputs):
    if all(x == 1 for x in inputs):
        return 1
    else:
        return -1
```

```python
def or_gate(inputs):
    if any(x == 1 for x in inputs):
        return 1
    else:
        return -1

def nand_gate(inputs):
    if all(x == 1 for x in inputs):
        return -1
    else:
        return 1

def nor_gate(inputs):
    if any(x == 1 for x in inputs):
        return -1
    else:
        return 1
```

```python
alpha=0.3
theta=1
```

```python
n=int(input("Enter the number of input:"))
```

```python
def generate_combinations(n):
    if n == 1:
        return [[1], [-1]]
    else:
        sub_combinations = generate_combinations(n - 1)
        combinations = []
        for sub_combination in sub_combinations:
            combinations.append(sub_combination + [1])
            combinations.append(sub_combination + [-1])
        return combinations
```

```python
combinations=generate_combinations(n)
```

```python
# Define your function to update weights
def update_weights(inputs, weights, bias, alpha, theta, select_gate):
    net_input = sum(x * w for x, w in zip(inputs, weights)) + bias
    if net_input >= theta:
        t = 1
    else:
        t = -1

    if select_gate == 'AND':
        y = and_gate(inputs)
    elif select_gate == 'OR':
        y = or_gate(inputs)
    elif select_gate == 'NAND':
        y = nand_gate(inputs)
    else:
        y = nor_gate(inputs)

    while t != y:
        for i in range(len(inputs)):
            weights[i] = weights[i] + alpha * y * inputs[i]
        bias = bias + alpha * y
        net_input = sum(x * w for x, w in zip(inputs, weights)) + bias
        if net_input >= theta:
            t = 1
        else:
            t = -1

    return weights, bias
```

```python
select_gate = 'AND'
```

```python
weights=[]
for i in range(n):
  k=int(input("Enter weights:"))
  weights.append(k)

```

```python
bias=int(input("Enter bias:"))
```

```python

for input_combination in combinations:
    weights, bias = update_weights(input_combination, weights, bias, alpha, theta, select_gate)

print("Final weights are:", weights)
print("Final bias is:", bias)
```
