```python
import pandas as pd
diabetes=pd.read_csv('drive/My Drive/Colab Notebooks/diabetes.csv')
```

```python
diabetes
```

```python
x=diabetes.iloc[:,0:8].values
x
```

```python
y=diabetes.iloc[:,8].values
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
x_train.shape
```

```python
#Standard scalar
from sklearn.preprocessing import StandardScaler
st=StandardScaler()
x_train=st.fit_transform(x_train)
x_test=st.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from sklearn import metrics
```

```python
classifier=Sequential()
```

```python
#Model 1 with 2 hidden layers
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=8))
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
classifier.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
classifier.fit(x_train,y_train,batch_size=10,epochs=10)
y_pred=classifier.predict(x_test)
y_pred=y_pred>0.5

prediction=metrics.accuracy_score(y_test,y_pred)
print("accuracy:",prediction *100,"%")
```

```python
#Model 2 with 3 hidden layers, kernel_initializer="he normal" and epochs=20
classifier.add(Dense(units=6,kernel_initializer='he_normal',activation='relu',input_dim=8))
classifier.add(Dense(units=6,kernel_initializer='he_normal',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='he_normal',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='he_normal',activation='relu'))
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
classifier.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
classifier.fit(x_train,y_train,batch_size=10,epochs=20)
y_pred=classifier.predict(x_test)
y_pred=y_pred>0.5

prediction=metrics.accuracy_score(y_test,y_pred)
print("accuracy:",prediction *100,"%")
```

```python
#Model 3 with 3 hidden layers, kernel_initializer="uniform" and epochs=20
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=8))
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
classifier.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
classifier.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
classifier.compile(optimizer="adam",loss='binary_crossentropy',metrics=['accuracy'])
classifier.fit(x_train,y_train,batch_size=10,epochs=20)
y_pred=classifier.predict(x_test)
y_pred=y_pred>0.5

prediction=metrics.accuracy_score(y_test,y_pred)
print("accuracy:",prediction *100,"%")
```
